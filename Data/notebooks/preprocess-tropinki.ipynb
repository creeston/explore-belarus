```python
from tinydb import TinyDB
db = TinyDB("../../tropinki.json")
len(db.all())
```

```python
places = db.all()
places[1]
```

```python
preprocessed_places = []
for place in places:
    if "place" not in place:
        continue
    location_parts = [l.strip() for l in place["place"]["address"].split(",")]
    last_part = location_parts[-1]
    if any(c.isdigit() for c in last_part) and len(location_parts) > 1 and len(last_part) < 10:
        location_parts[-2] = location_parts[-2] + " " + last_part
        location_parts = location_parts[:-1]
    preprocessed_places.append({
        "name": place["title"],
        "url": {'href': place["url"], 'text': place['title']},
        "image": place["image"],
        "location": location_parts,
        "coords": place["place"]["coordinates"]
    })
preprocessed_places[0]
```

```python
for place in preprocessed_places:
    print(place["location"])
```

```python
import json

with open("tropinki_processed.json", "w", encoding="utf-8") as f:
    json.dump(preprocessed_places, f, ensure_ascii=False, indent=2)
```
