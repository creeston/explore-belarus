```python
from tinydb import TinyDB
db = TinyDB("../../vedaj_ru.json")
len(db.all())
```

```python
places = db.all()
places[1]
```

```python
for place in places:
    print(place["place"]["location"])
```

```python
preprocessed_places = []
for place in places:
    # check if name contains string in parantheses
    # then extract part in parathenses and remove it from the string
    place_name = None
    former_name = None
    if "(" in place["name"]:
        name = place["name"]
        start = name.find("(")
        end = name.find(")")
        place_name = name[:start] + name[end+1:].replace("  ", " ").strip()
        former_name = name[start+1:end].strip()
    else:
        place_name = place["name"]


    location = tuple(place["place"]["location"])
    if location == ("Город Минск",):
        location = ('Минская область', "Минск")

    preprocessed_places.append({
        "name": place_name,
        "former_name": former_name,
        "location": location,
        "images": place["place"]["images"],
        "url": {'href': place["url"], 'text': place['name']},
        "tags": place["place"]["tags"],
    })
preprocessed_places[0]
```

```python
import json

with open("vedaj_processed.json", "w", encoding="utf-8") as f:
    json.dump(preprocessed_places, f, ensure_ascii=False, indent=2)
```
